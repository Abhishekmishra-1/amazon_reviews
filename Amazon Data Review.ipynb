```python
import seaborn as sns
import pandas as pd
%matplotlib inline

"""Text Token Heatmap: darker colors imply more predicitve"""
'''
In this graph we grouped the overall sentiment that AWS Comprehend 
assigned to a given review and placed it inbuckets by ratings. 
This was done both for the text of reviews and their summary.

As the two heatmaps below reveal, the summary records that are positive do tend to cluster more strongly towards
higher ratings, but negative ratings do not reveal a similar bias.

The main takeaway here is that low scores to not strongly correlate to the way the AWS Comprehend identifies
negative sentiment.
'''
df = pd.read_pickle('/Users/drewmassey/git-repos/amazon_reviews/data/summary_token_count.pickle')
sns.heatmap(df)
```

```python
"""Summary Token Heatmap: darker colors imply more predictive."""
df = pd.read_pickle('/Users/drewmassey/git-repos/amazon_reviews/data/text_token_count.pickle')
sns.heatmap(df)
```

```python
"""Correlation analysis between texts and summarys"""
'''
This chart looks to see if there is any chance that the text and summary might yield similar sentiments.
What I did was build an intermediate data set which calculated the euclidean distance of the sentiment 
vectors returned by AWS Comprehend, and then grouped them by the rating.

As this chart shows, the data is scattered a bit too broadly to be useful in this approach.
'''
delta_analysis = pd.read_csv('/Users/drewmassey/git-repos/amazon_reviews/data/text_summary_delta.csv', header=None)
delta_analysis.columns = ['delta','rating']
delta_analysis = delta_analysis.sample(frac=0.005)

delta_analysis.plot(kind='scatter', x='rating', y='delta')
# delta_analysis.plot()
```
